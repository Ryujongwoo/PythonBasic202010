튜플 만들기  
튜플은 리스트와 비슷하지만 리스트는 []로 둘러싸고 튜플은 ()로 둘러싼다는 것과 리스트는 데이터의 수정, 삭제가 가능하지만 튜플은 수정, 삭제가 불가능하다.

```python
# 빈 튜플 만들기
a = ()
print(a)
print(type(a))
# 생성자 함수를 이용해 빈 튜플 만들기
b = tuple()
print(b)
print(type(b))
```

```python
c = [1]
print(c)
print(type(c))
d = (1) # ()를 일반적인 연산식에 사용하는 ()로 인식해서 상수를 기억하는 변수가 된다.
print(d)
print(type(d))
e = (1,) # 데이터를 1개만 가지는 튜플을 만들어야 한다면 반드시 데이터 뒤에 ','를 찍어야 한다.
print(e)
print(type(e))
```

```python
f = (1, 2)
print(f)
print(type(f))
g = (1, 2, 3)
print(g)
print(type(g))
h = (1, 2, ('a', 'b'))
print(h)
print(type(h))
i = (1, 2, ['a', 'b'])
print(i)
print(type(i))
j = [1, 2, ('a', 'b')]
print(j)
print(type(j))
```

```python
# 튜플은 ()를 생략해도 변수에 데이터를 ','로 구분해서 여러개 넣어주면 자동으로 튜플로 인식한다.
k = 1, 2, 3, 4, 5, 6, 7
print(k)
print(type(k))
```

```python
# 튜플은 리스트와 같이 인덱싱과 슬라이싱이 가능하다.
l = 1, 2, 3, 4, 5, 6, 7
print(l)
print(l[0])
print(l[-1])
print(l[2:4])
print(l[:4])
print(l[2:])
# 튜를은 리스트와 같이 '+', '*' 연산이 가능하다.
m = 'a', 'b'
print(m)
print(m + l)
print(m * 3)
```

```python
# 튜플은 리스트와 달리 수정과 삭제를 할 수 없다.
n = 1, 2, 3, 4, 5, 6
print(n)
# n[0] = 'z' # 튜플에 저장된 데이터를 수정하려 했으므로 에러가 발생된다.
# del n[0]   # 튜플에 저장된 데이터를 삭제하려 했으므로 에러가 발생된다.
del n        # del 명령으로 튜플 자체를 삭제하는 것은 가능하다.
print(n)
```

```python
# 두 기억장소에 저장된 데이터 바꾸기
a = 3
b = 4
print('a = {}, b = {}'.format(a, b))

temp = a
a = b
b = temp
print('a = {}, b = {}'.format(a, b))

[a, b] = [b, a]
print('a = {}, b = {}'.format(a, b))

a, b = b, a
print('a = {}, b = {}'.format(a, b))
```
