함수의 구조  
def 함수이름([인수, ...]):  
&nbsp;&nbsp;&nbsp;&nbsp;함수가 실행할 문장  
&nbsp;&nbsp;&nbsp;&nbsp;...
&nbsp;&nbsp;&nbsp;&nbsp;[return [함수의 실행 결과 값]]  

[]로 묶어놓은 함수의 인수와 return은 필요 없으면 생략이 가능하다.

함수의 실행 순서  
프로그램에서 함수가 호출되면 호출된 함수로 제어가 넘어간다. => 복귀 주소는 스택에 저장된다.  
호출된 함수에서 정의한 기능을 모두 실행하거나 return을 만나면 함수가 호출된 것으로 스택에 저장된 복귀 주소를 참조해서 돌아간다. 
return 뒤에 값이 있으면 값을 가지고 호출된 곳으로 돌아가고 값이 없으면 None을 가지고 돌아간다.

```python
# 가장 일반적인 함수의 모양
# a, b 2개의 데이터를 인수로 넘겨받아 덧셈을 실행한 후 결과를 return하는 함수
def add(a, b):
    print('add(a, b) 함수가 실행됩니다.')
    result = a + b
    a = 999
    print('가인수 a : {}'.format(a))
    return result
```

실인수와 가인수는 독립된 메모리 공간을 사용하기 때문에 가인수의 값이 변경되더라도 실인수의 값은 변경되지 않는다.  
실인수와 가인수의 이름은 반드시 같아야 할 필요는 없지만 개수는 반드시 같아야 한다.

```python
a, b = map(int, input('정수 2개를 입력하세요 : ').split())
print(add(a, b))
print('실인수 a : {}'.format(a))
```
